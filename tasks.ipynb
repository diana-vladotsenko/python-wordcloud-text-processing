## Task 1. 

## Explanation
The code downloads a text file from Project Gutenberg and turns it into a string. It shows a short preview by taking the first 200 characters. The text is then split into a list of words. The total number of words is counted. A list of longer words is created by selecting words with more than six letters. Another list collects the first letter of words that have more than three characters. Finally, the word list is reversed so you can see the last part of the text.


```python
import urllib.request

text = urllib.request.urlopen(
    "https://www.gutenberg.org/cache/epub/11/pg11.txt"
).read().decode("utf-8")

preview = text[:200]

words = text.split(" ")

word_count = len(words)

long_words = [word for word in words if len(word) > 6][:20]

first_letters = [word[0] for word in words if len(word) > 3][:20]

reversed_words = words[::-1][:15]

preview, words[:10], word_count, long_words, first_letters, reversed_words
```

## Task 2. 

## Explanation
This code generates two word clouds from the same text. The first word cloud uses the default settings and shows which words appear most often in the text, so the most frequent words appear larger. The second word cloud removes common stopwords like “and”, “the”, “to”, “of”, and “in”. After filtering these out, more meaningful content words become more visible in the image.



```python
from wordcloud import WordCloud

wc_basic = WordCloud(
    background_color="white"
).generate(text)

basic_img = wc_basic.to_image()
basic_img.show()

custom_stopwords = ["and", "the", "to", "of", "in"]

wc_filtered = WordCloud(
    background_color="white",
    stopwords=custom_stopwords
).generate(text)

filtered_img = wc_filtered.to_image()
filtered_img.show()
```

The largest words in the cloud are “said”, “Alice”, It, “little”, “that”, and “you”, which means these appear most frequently in the text. This shows that the story is centered around Alice and her conversations, since dialogue words like “said” dominate the cloud. After removing common stopwords, more meaningful words related to the characters and events become visible, which helps highlight the themes and recurring names in the text.
